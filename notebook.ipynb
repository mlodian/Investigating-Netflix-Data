```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Start coding!
```

```python
# Created a DataFram from the csv "netflix_data"
netflix_df = pd.read_csv("netflix_data.csv")
# Inspect by viewing the first 5 rows using head
print(netflix_df.head())
```

```python
# Subset the DataFrame by choosing the row where the type is a "Movie"
netflix_subset = netflix_df[netflix_df['type'] == 'Movie']
```

```python
# Now to make the columns of the new DataFrame
netflix_movies = netflix_subset[['title', 'country', 'genre', 'release_year', 'duration']]
```

```python
# Lest make a short movies list for movies shorter than 60 mins
short_movies = netflix_movies[netflix_movies['duration'] < 60]
print(short_movies.head(20))
```

```python
# Imma add colors to this
colors = []

# Iterating netflix_subset
for genre in netflix_movies['genre']:
    if genre == "Children":
        colors.append('blue')
    elif genre == "Documentaries":
        colors.append('green')
    elif genre == "Stand-Up":
        colors.append('red')
    else:
        colors.append('black')
```

```python
import matplotlib.pyplot as plt

# Let's create a scatter plot
fig = plt.figure(figsize=(12, 8))
plt.scatter(netflix_movies['release_year'], netflix_movies['duration'], c=colors)

# Now with add labels and title
plt.xlabel('Release year')
plt.ylabel('Duration (min)')
plt.title('Movie Duration by Year of Release')

# The colors will be shown now
plt.show()
```
